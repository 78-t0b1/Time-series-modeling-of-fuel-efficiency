# Auto Corelation Plot 

instead of hypothesising autocorelation in data we can easily plot autocorelation of data with varrying window size to examine seasonality and also determining if its stationary or not.

plot_acf is used to plot ACF plot

if we se gradual decline with seasonal pattern then data is corelated and also non stationary.But if we observe sudden drop in corelarion then data is stationary. Shaded region indicates 95% confidance interval.


```python
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf

fig,ax = plt.subplots(len(eng_list),figsize=(20,5))
i = 0
plt.subplots_adjust(bottom=0.1, right=0.8, top=10.9)
for x in eng_list:
    g = grouped.get_group(x)
    plot_acf(g['operating_hours'],ax=ax[i]);
#     plot_acf(g['fuel_consumption(liter)'],ax=ax[i]);
    i+=1
```

   # ARIMA (Auto Reg. Integration Moving Average)
   
   non seasonal arima has threee components p,d,q which are non-negative integers.
   
   AR(p) : A regression model that utilizes dependent relationship between a current observation and observations over previous period.
   
   
   I(d) : Diffrencing of observations (y(t)-y(t-1)) in order to make the time series stationary.( how many times we have to diffrence the dataset to make it stationary.)
   
   
   MA(q) : A model that uses the dependency between an observation and a residual error from a moving average model applied to lagged observations.
   

###  A Stationary data has constant mean and variance over time.  Mean needs to be constant in stationary and variance or co-variance should not be a function of time.

## Auto Regression Models

In an autoregression model we forecast using linear combination of past values of variable. An autoregression is run against a set of laggeed values of order P.The term auto-reg. describes a regression of variable against itself.

The auto-reg model specifies that output variable depends linearly on its previous values and on an imperfectly preedictable term(stochastic term). 



```python
from statsmodels.tsa.ar_model import AR,ARResults
import warnings
warnings.filterwarnings('ignore')
```

```python
g = grouped.get_group(2582458400)
g.head()
```

```python
train = g.iloc[:80]
test = g.iloc[80:]
```

```python
ar_model = AR(train['operating_hours'])
```

```python
ar1_fit = ar_model.fit(maxlag=1)
```

```python
ar1_fit.params
```

```python
start = len(train)
stop = len(g)-1
```

```python
pre1 = ar1_fit.predict(start=start,end=stop)
```

```python
pre1 = pre1.rename('AR1 predictions')
```

```python
test['operating_hours'].plot(figsize=(20,5),legend=True)
pre1.plot(figsize=(20,5),legend=True)
```

```python
ar20_fit = ar_model.fit(maxlag=20)
pre20 = ar20_fit.predict(start=start,end=stop)
pre20 = pre20.rename('AR20 predictions')
```

```python
test['operating_hours'].plot(figsize=(12,5),legend=True)
pre1.plot(figsize=(12,5),legend=True)
pre20.plot(figsize=(12,5),legend=True)
```
